# import libraries and data

```python
import numpy as np
import pandas as pd

df=pd.read_csv('D:\\Datasets\\Housingdata.csv')
df
```

```python
x=df.iloc[:,1:6]
x
y=df.iloc[:,0:1]
print(x,y)
```

# Split data and fit model 

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor


x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,test_size=0.20,random_state=2018)

knn=KNeighborsRegressor(n_neighbors=5)

knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```
