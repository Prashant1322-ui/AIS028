```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:\\Datasets\\Fruits color datasets.csv")
df
```

```python
#checking how many unique fruit names are present in the dataset
df['fruit_name'].value_counts()
```

```python
x=df.iloc[:,3:6]
y=df.iloc[:,0:1]
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=10)
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
x_train.describe()
```

```python
x_test.describe()
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)

knn.fit(x_train,y_train)
```

```python
#prediction
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
c_m=confusion_matrix(y_test,y_pred)
c_m
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#example1 parameters of following function are mass,width and height
y1=knn.predict([['100','6.3','8']])
y1[0]
```

```python
y2=knn.predict([['170','7.5','9']])
y2[0] # zero means label fruits type 1,2,3 other error 
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DTK=DecisionTreeClassifier()
DTK.fit(x_train,y_train)
```

```python
y_pred=DTK.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn import tree
plt.figure(figsize=(16,9))
tree.plot_tree(DTK,filled=True)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=100,random_state=0)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
c_m=confusion_matrix(y_test,y_pred)
c_m
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
