```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df
```

```python
x=df.iloc[:,0:4]
y=df.iloc[:,4:5]
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test=train_test_split(x,y,train_size=0.80,random_state=101)

print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```
